```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
ar = range(0,180)
ar2 = range(180,0,-1)

values = []
values2 = []

base = 50
scale = 10

for a in ar:
    #values.append([a, abs(a-90)/3+15])
    values.append([a, base+math.sin(math.radians(a)*2)*scale])

for a in ar2:
    values2.append([a, base+-math.sin(math.radians(a)*2)*scale])
    #values2.append([a, 180-abs(a-90)/3-105])
    
values = np.array(values)
values2 = np.array(values2)

plt.plot(values[:,0], values[:,1])
plt.plot(values2[:,0], values2[:,1])
plt.ylim((0,180))
```
